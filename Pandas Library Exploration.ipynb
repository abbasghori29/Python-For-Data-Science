# Clean Messy Values From Data set By strip:

```python
import pandas as pd

# Create a sample DataFrame with a column containing strings with leading and trailing whitespace
data = {'Text': ['Hello@@@', '@@@World', 'Python@@@', '@@@Pandas']}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Use str.strip() to remove leading and trailing whitespace from the 'Text' column
df['Text'] = df['Text'].str.strip("@@@")

# Display the DataFrame after stripping whitespace
print("\nDataFrame after stripping whitespace:")
print(df)
```
