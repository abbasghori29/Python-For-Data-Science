# Side By Side Bars

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = [10, 15, 20, 25, 30]
data2 = [8, 12, 18, 22, 28]

x = np.arange(len(data1))
bar_width = 0.35

plt.bar(x - bar_width/2, data1, bar_width, label='Data 1')
plt.bar(x +bar_width/2, data2, bar_width, label='Data 2')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Side-by-Side Bar Chart')
plt.xticks(x, ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5'])
plt.legend()

plt.show()
```

# Axis Of Line Graph (zoom to axis which you want to show)

```python
import matplotlib.pyplot as plt

x = [33, 44, 50, 70]
y = [55, 80, 65, 60]
z = [55, 8, 45, 8]

plt.plot(x, y, marker="o")
plt.plot(x, z, marker="*")

# Calculate axis limits based on data range
xmin, xmax = min(x), max(x)
ymin, ymax = min(min(y), min(z)), max(max(y), max(z))

plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# Create a date range for transactions
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Generate random data for sales transactions
num_transactions = 1000
regions = ['Region A', 'Region B', 'Region C']
products = ['Product1', 'Product2', 'Product3', 'Product4', 'Product5']
data = {
    'Date': random.choices(date_range, k=num_transactions),
    'Region': random.choices(regions, k=num_transactions),
    'ProductID': random.choices(products, k=num_transactions),
    'QuantitySold': np.random.randint(1, 100, size=num_transactions),
    'TotalSalesAmount': np.random.uniform(100, 1000, size=num_transactions)
}
df=pd.DataFrame(data)
# df=df.groupby("ProductID")["TotalSalesAmount"]
top5=df.nlargest(5,"TotalSalesAmount")
top5.sort_values("TotalSalesAmount",ascending=False)
top5
plt.bar(top5["TotalSalesAmount"],top5["Region"])
plt.show()
# Grouping by "ProductID" and aggregating to sum the quantity sold for each product
product_totals = df.groupby("ProductID").agg({"QuantitySold": "count"}).reset_index()
product_totals
```

# Subplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a sample DataFrame
data = {
    'Customer_ID': [1, 2, 3, 4, 5],
    'Age': [25, 30, 35, 40, 45],
    'Purchase_Frequency': [10, 5, 8, 12, 7],
    'Product_Category_1': [2, 1, 3, 2, 1],
    'Product_Category_2': [3, 2, 1, 2, 1],
    'Product_Category_3': [1, 3, 2, 1, 2]
}
df = pd.DataFrame(data)

# Creating subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Subplot 1: Scatter Plot
axes[0, 0].scatter(df['Age'], df['Purchase_Frequency'])
axes[0, 0].set_title('Age vs Purchase Frequency')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Purchase Frequency')

# Subplot 2: Histogram
axes[0, 1].hist(df['Product_Category_1'])
axes[0, 1].set_title('Product Category 1 Distribution')
axes[0, 1].set_xlabel('Product Category 1')
axes[0, 1].set_ylabel('Frequency')

# Subplot 3: Boxplot
sns.boxplot(x=df['Product_Category_2'], ax=axes[1, 0])
axes[1, 0].set_title('Product Category 2 Boxplot')
axes[1, 0].set_xlabel('Product Category 2')
axes[1, 0].set_ylabel('Values')

# Subplot 4: Heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', ax=axes[1, 1])
axes[1, 1].set_title('Correlation Matrix')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
