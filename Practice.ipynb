```python
from abc import ABC,abstractmethod
class Shape(ABC):
    def __init__(self,length):
        self.length=length;
    @abstractmethod
    def getArea(self):
        pass
    
class Circle(Shape):
    def __init__(self,length,naam):
        super().__init__(length)
        self.naam=naam;
    def getArea(self):
        print(f"Area Of {self.naam} = {3.1424*self.length**2}")
        
class Square(Shape):
    def __init__(self,length,name):
        super().__init__(length)
        self.name=name;
    def getArea(self):
        print(f"Area Of {self.name} = {self.length**2}")
c=Circle(9,"GKB")
s=Square(4,"BKB")
c.getArea()   
s.getArea()  
```

```python
class student:
    def __init__(self,iD,name):
        self.iD=iD
        self.name=name
    def getInfo(self):
        print(f"Name Of Student Is {self.name} and his ID is {self.iD}")
class Marks(student):
    def __init__(self,iD,name,marksinpf,marksinoop):
        super().__init__(iD,name)
        self.marksinpf=marksinpf
        self.marksinoop=marksinoop
    def getmarksdetail(self):
        print(f"{self.marksinpf}")
        print(f"{self.marksinoop}")
class Result(Marks):
    def __init__(self,iD,name,marksinpf,marksinoop):
        super().__init__(iD,name,marksinpf,marksinoop)
    def getTotalMarks(self):
        print(f"{self.marksinpf+self.marksinoop}")

r=Result("22K-4026","Abbas",86,87)
r.getInfo()
r.getmarksdetail()
r.getTotalMarks
        
```
